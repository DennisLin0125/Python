```python
#獲取版本
import pandas as pd
pd.__version__
```

```python
#用 Python dict 建立 DataFrame
'''
基本上 dict 裡頭的每一個鍵值（key）都對應到一個欄位名稱，而其值（value）則是一個 
iterable，代表該欄位裡頭所有的數值。
'''
dic = {
    "col 1": [1, 2, 3], 
    "col 2": [10, 20, 30],
    "col 3": list('xyz'),
    "col 4": ['a', 'b', 'c'],
    "col 5": pd.Series(range(3))
}
df = pd.DataFrame(dic)
df
```

```python
#改變 DataFrame 裡的欄位名稱
rename_dic = {"col 1": "x", "col 2": "10x"}
df.rename(rename_dic, axis=1)
```

```python
#用 df.columns 的方式改欄位名稱：
df.columns = ['x(new)', '10x(new)'] + list(df.columns[2:])
df
```
